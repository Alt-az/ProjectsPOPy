```python
import matplotlib.pyplot as plt
import numpy as np
import skimage
img=skimage.data.chelsea() 
histb, bin_edges = np.histogram(img[:,:,2],bins=range(257))
histg, bin_edges = np.histogram(img[:,:,1],bins=range(257))
histr, bin_edges = np.histogram(img[:,:,0],bins=range(257))
x=np.arange(0, 256, 1)
plt.plot(x,histb,'b',label = "niebieski")
plt.plot(x,histg,'g',label = "zielony")
plt.plot(x,histr,'r',label = "czerwony")
plt.legend()
plt.show()
fig = plt.subplots(figsize =(12, 8))
histb, bin_edgesb = np.histogram(img[:,:,2],bins=8)
histg, bin_edgesg = np.histogram(img[:,:,1],bins=8)
histr, bin_edgesr = np.histogram(img[:,:,0],bins=8)
barWidth=0.25
x1=np.arange(0, 8, 1)
x2=[x + barWidth for x in x1]
x3=[x + barWidth for x in x2]
plt.bar(x3,histb,color ='b',width=barWidth,label = "niebieski")
plt.bar(x2,histg,color ='g',width=barWidth,label = "zielony")
plt.bar(x1,histr,color ='r',width=barWidth,label = "czerwony")
plt.xlabel('Zakresy', fontweight ='bold', fontsize = 15)
plt.ylabel('Piksele', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(8)],
        ['(0,32)', '(32,64)', '(64,96)', '(96,128)', '(128,160)','(160,192)','(192,224)','(224,256)'])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import data
from skimage import exposure
import cv2
def fun(x,a):
    b=-128*a+128
    return int((x*a)+b)
img=data.camera()
a=5
fig = plt.subplots(figsize =(12, 8))
x=np.arange(0, 256, 1)
y=np.array([fun(x,a) for x in range (0,256)]).clip(0,255).astype('uint8')
y[y<0]=0
y[y>255]=255
plt.plot(x,y)
plt.show()
LUT = np.array([fun(x,a) for x in range (0,256)]).clip(0,255).astype('uint8')
img2=cv2.LUT(img,LUT)
plt.imshow(img2,cmap='gray')
plt.show()
```
